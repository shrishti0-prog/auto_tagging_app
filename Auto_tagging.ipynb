```python
import re

def basic_clean(text):
    text = re.sub(r"<.*?>", " ", text)   # remove HTML tags (FAST)
    text = re.sub("[^a-zA-Z]", " ", text)
    text = text.lower()
    text = re.sub(r"\s+", " ", text)
    return text

questions_df["cleaned_text"] = questions_df["Body"].fillna("").apply(basic_clean)
```

```python
df=pd.merge(questions_df, tags_df, how='inner',on='Id')
```

```python
df=df[['Id','Body','cleaned_text','tags']]
print("shape=>", df.shape)
df.head()
```

```python
freq={}
for i in df['tags']:
    for j in i:
        if j in freq.keys():
          freq[j]=freq[j]+1
        else:
            freq[j]=1
```

```python
def classify(pred_prob, threshold):
    return (pred_prob >= threshold).astype(int)
```

```python
import numpy as np
from sklearn.metrics import f1_score

def optimum_threshold(actual, pred_prob):
    thresholds = np.arange(0, 0.5, 0.1)  # from 0 to 0.5

    score = []
    
    for value in thresholds:
        pred_classes = classify(pred_prob, value)
        score.append(f1_score(actual, pred_classes, average="weighted"))
        
    return thresholds[score.index(max(score))]
```

```python
print("Optimal Threshold=>", optimum_threshold(y_tr, train_pred_nb))
```

```python
def tagging(question):
    cleaned_question=basic_clean(question)
    vector=word_vectorizer.transform([cleaned_question])
    pred_prob=lr_model.predict_proba(vector)
    pred_class=classify(pred_prob,0.300004)

    return mlb.inverse_transform(np.array(pred_class))
    
```

```python
import numpy as np
all_tags = mlb.classes_
def tagging(question, top_k=3):
    cleaned_question = basic_clean(question)

    vector = word_vectorizer.transform([cleaned_question])
    pred_prob = lr_model.predict_proba(vector)

    probs = pred_prob[0]

    top_indices = np.argsort(probs)[-top_k:]
    tags = [all_tags[i] for i in top_indices]

    return tags
```
